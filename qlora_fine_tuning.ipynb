````python
from unsloth import FastLanguageModel
import torch

print(f"GPU : {torch.cuda.get_device_name(0)}")
print(f"VRAM : {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} Go")
print("✅ Unsloth OK !")
````

````python
from unsloth import FastLanguageModel

max_seq_length = 2048
load_in_4bit = True

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    max_seq_length = max_seq_length,
    dtype = None,
    load_in_4bit = load_in_4bit,
)

# Configuration LoRA
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],
    lora_alpha = 32,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 42,
)

model.print_trainable_parameters()
````

````python
from datasets import load_dataset
from unsloth.chat_templates import get_chat_template, train_on_responses_only

# Appliquer le template ChatML au tokenizer
tokenizer = get_chat_template(
    tokenizer,
    chat_template = "chatml",
)

# Fonction qui formate chaque exemple en texte ChatML
def formatting_prompts_func(examples):
    convos = examples["messages"]
    texts = [
        tokenizer.apply_chat_template(
            convo,
            tokenize = False,
            add_generation_prompt = False
        )
        for convo in convos
    ]
    return {"text": texts}

# Charger ton dataset
dataset = load_dataset("json", data_files="/content/loodo_dataset.jsonl", split="train")

# Appliquer le formatage
dataset = dataset.map(formatting_prompts_func, batched=True)

# Vérification — affiche le premier exemple formaté
print("Nombre d'exemples :", len(dataset))
print("\n--- Premier exemple formaté ---")
print(dataset[0]["text"][:500])
````

````python
from transformers import TextStreamer

messages_test = [
    {"role": "system", "content": "Tu es Loodo, un coach bienveillant spécialisé en motivation et organisation personnelle."},
    {"role": "user", "content": "J'arrive pas à me concentrer depuis 2 heures, je procrastine sur tout."}
]

inputs = tokenizer.apply_chat_template(
    messages_test,
    tokenize = True,
    add_generation_prompt = True,
    return_tensors = "pt"
).to("cuda")

streamer = TextStreamer(tokenizer, skip_prompt=True)

print("=== RÉPONSE AVANT FINE-TUNING ===\n")
with torch.no_grad():
    _ = model.generate(
        input_ids = inputs,
        streamer = streamer,
        max_new_tokens = 200,
        temperature = 0.7,
        do_sample = True,
        use_cache = False,  # ← on désactive le cache KV
    )
````

Réponces du  model Mistral-07 avent Fine tuning:
Cela peut être dû à plusieurs facteurs, mais l'une des causes les plus courantes est l'anxiété chronique. Vous pouvez essayer de vous poser des questions telles que "Qu'est-ce qui m'empêche de me concentrer ?", "Quelle est la pire chose qui peut arriver si je ne suis pas distrait maintenant ?" et "Quand je suis concentré, je peux réussir à accomplir ce qui est important pour moi. Quand je suis distrait, je me sens perdu et je n'ai pas de réussite". En répondant à ces questions, vous pouvez comprendre les causes de votre distraction.<|im_end|>


````python
from trl import SFTTrainer
from transformers import TrainingArguments

# Reformatage manuel avec les bons tokens
def formatting_func(example):
    text = ""
    for msg in example["messages"]:
        role = msg["role"]
        content = msg["content"]
        text += f"<|im_start|>{role}\n{content}<|im_end|>\n"
    text += "<|im_start|>assistant\n"
    return {"text": text}

dataset_fixed = dataset.map(formatting_func)

# Vérifie que ça ressemble à ça
print(dataset_fixed[0]["text"][:300])
````

````python
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset_fixed,
    dataset_text_field = "text",
    max_seq_length = 2048,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        num_train_epochs = 3,
        learning_rate = 2e-4,
        fp16 = True,
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 42,
        output_dir = "loodo_outputs",
    ),
)

trainer_stats = trainer.train()

print(f"\n✅ Training terminé !")
print(f"Loss finale : {trainer_stats.training_loss:.4f}")
````

✅ Training terminé !
Loss finale : 0.9451

````python
messages_test = [
    {"role": "system", "content": "Tu es Loodo, un coach bienveillant spécialisé en motivation et organisation personnelle."},
    {"role": "user", "content": "J'arrive pas à me concentrer depuis 2 heures, je procrastine sur tout."}
]

inputs = tokenizer.apply_chat_template(
    messages_test,
    tokenize = True,
    add_generation_prompt = True,
    return_tensors = "pt"
).to("cuda")

streamer = TextStreamer(tokenizer, skip_prompt=True)

print("=== RÉPONSE APRÈS FINE-TUNING ===\n")
with torch.no_grad():
    _ = model.generate(
        input_ids = inputs,
        streamer = streamer,
        max_new_tokens = 200,
        temperature = 0.7,
        do_sample = True,
        use_cache = False,
    )
````

réponce après fine tuning :

Procrastination c'est souvent le signe que le cerveau a besoin de direction. On va créer un plan d'action de 2 minutes, pas un plan de 2 heures — le cerveau ne s'engage pas avec ça. Dis-moi ce que tu dois faire aujourd'hui en 3 mots maximum. Ensuite, on découpe ce incompletable en 3 à 5 actions concrètes de 5 minutes chacune. On part de l'imprévisible au prédicable. Par exemple : 'Rédiger le mémo pour la réunion demain' devient 'Ouvre Google Docs, écrit le titre, développe le point 1 en 2 lignes, enregistre et ferme — 5 minutes.' Le cerveau se dit 'je peux ça'.<|im_end|>

Deux changements par rapport à v1 : learning_rate baissé à 1e-4 car le modèle a déjà appris une base, et warmup_steps monté à 10 pour le dataset plus grand.

Reponse de la V2

2 heures de perte de concentration, c'est souvent le signal que ton cerveau a besoin d'une pause. Pas d'accuse, pas de culpabilité — tu as travaillé avant. Fais une micro-pause de 5 minutes : marche, station debout, stretching, respiration. Puis relance sur une micro-tâche précise de moins de 5 minutes. Travaller fatigué te coûte plus de temps au final.<|im_end|>


Pourquoi il dit "tu as travaillé avant" :
Le modèle ne sait pas que tu procrastines depuis 2h sans avoir travaillé. Il a appris dans son dataset que quand quelqu'un dit "je n'arrive pas à me concentrer", c'est souvent après une période de travail. Il hallucine une hypothèse basée sur les patterns statistiques de tes données d'entraînement — pas sur la réalité de la conversation.
C'est une limite fondamentale du fine-tuning sur peu de données : le modèle généralise parfois trop vite et invente du contexte.
